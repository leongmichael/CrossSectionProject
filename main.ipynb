```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
import math
```

```python
x = np.linspace(-8,8,100) # Return evenly spaced numbers over a specified interval 

# list of equations to plot. add as many as you need
def f(x): return -0.002*x**4 + 8
def g(x): return -0.002*x**4 + 4
```

```python
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.axhline(color = 'black')
plt.xlim([-8, 8])
plt.ylim([0, 8.5])

```

```python
start_bound = -7.953
end_bound = 7.953
increment = 0.5 # thickness of each slice 

vol_slices = 0

x = sy.Symbol('x')
#total volume
print("Exact Volume", sy.integrate((np.pi)/8 * (f(x) - g(x))**2, (x,start_bound, end_bound)))

for i in np.arange(start_bound, end_bound, increment):
    vol_slices += sy.integrate((np.pi)/8 * (f(x) - g(x))**2, (x,start_bound, start_bound + increment))
    # print("Start Bound:", i, "End Bound:", i + increment, "Volume:", sy.integrate((np.pi)/8 * (f(x) - g(x))**2, (x,start_bound, start_bound + increment)))

print("Total Slices:", math.ceil((abs(start_bound*2) / increment)), " whole slices /", (abs(start_bound*2) / increment), "exact slices")

print("Volume from Slices:", vol_slices)
```

```python
print("Start Bound, End Bound, Diameter, Area, Volume")
x = sy.Symbol('x')
for i in np.arange(start_bound, end_bound, increment):
    print(i, #start bound
          i + increment, # end bound 
          f(start_bound)- g(start_bound), # diameter
          ((np.pi / 2) * ((f(start_bound)- g(start_bound))/2)**2), #area
          sy.integrate((np.pi)/8 * (f(x) - g(x))**2, (x,start_bound, start_bound + increment)) #volume
          )
```
